# Step-3

1. Imputation on Embarked and Age columns

By imputation, we mean to replace the missing or null values with a particular value in the entire dataset. Imputation can be done using any of the below techniques– Impute by mean. Impute by median.
1. Filling the missing data with the mean or median value if it’s a numerical variable.
2. Filling the missing data with mode if it’s a categorical value.

Embarked column is a Categorical column so we impuate by the mode

```python
data['Embarked'].value_counts() 
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode())
```

```python
data['Embarked'].fillna('S',inplace =True)
```

```python
data
```

```python
data.isnull().sum()
```

Age column is a numerical column so we impuate by the mean.

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data[data["Age"].isnull()]
```

```python
data.isnull().sum()
```

6. one hot encoding

```python
data['Survived'] = data['Survived'].astype('category')
data['Pclass'] = data['Pclass'].astype('category')
data['Sex'] = data['Sex'].astype('category')
data['SibSp'] = data['SibSp'].astype('category')
data['Sex'] = data['Sex'].astype('category')
data['Parch'] = data['Parch'].astype('int')
data['Embarked'] = data['Embarked'].astype('category')
```

```python
data
```

```python
columns = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']
for col in columns:
    data= pd.concat([data, pd.get_dummies(data[col], drop_first = True)], axis =1)
```
